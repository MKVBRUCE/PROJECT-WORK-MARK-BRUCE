```python
import pandas as pd
plt.figure(figsize=(9,7))
dataset=pd.read_csv('red_wine.csv')
plt.hist(dataset['alcohol'],bins=20,color='red', edgecolor='yellow')
plt.title('ILLUSTRATION OF THE ALCOHOLIC CONTENT IN RED WINE')
plt.xlabel ('ALCOHOLIC CONTENT')
plt.ylabel ('Frequency')
plt.show()
         
         
```

```python
plt.figure(figsize=(9,7))
plt.scatter(dataset['alcohol'], dataset['fixed acidity'], color='blue', edgecolor='black',)
plt.title('ALCOHOLIC CONTENT AGAINST FIXED ACIDITY IN RED WINE')
plt.xlabel('Alcoholic content')
plt.ylabel('fixed acidity')
plt.show()
          
```

```python
plt.figure(figsize=(8,6))
plt.hist(dataset['citric acid'],bins=20,color='yellow',edgecolor='black')
plt.title('Illustration of amount of citric acid in red wine')
plt.xlabel('Citric Acid Content')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(9,7))
plt.scatter(dataset['residual sugar'],dataset['sulphates'],color='brown')
plt.title('RESIDUAL SUGAR CONTENT AGAINST SULPHATES IN RED WINE')
plt.xlabel('Residual sugar content')
plt.ylabel('Sulphates')
plt.show()
```

```python

import matplotlib.pyplot as plt
plt.figure
plt.figure(figsize=(8,6))
import pandas as pd
pd
dataset=pd.read_csv('red_wine.csv')
plt.scatter(dataset['total sulfur dioxide'],dataset['pH'],color='orange')
plt.title('RELATIONSHIP OF TOTAL SULFUR DIOXIDE WITH pH of Red wine')
plt.xlabel('Total sulfur dioxide')
plt.ylabel('pH of red wine')
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.hist(dataset['volatile acidity'], bins=20, color='green', edgecolor='black')
plt.title('Pictorial Illustration of Volatile acidity in red wine')
plt.xlabel('Volatile Acidity')
plt.ylabel('frequency')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(dataset['density'],dataset['quality'],color='black')
plt.title('Relationship between density and quality of red wine')
plt.xlabel('Density')
plt.ylabel('quality')
plt.show()

```

```python
plt.figure(figsize=(8,6))
plt.hist(dataset['free sulfur dioxide'], bins=20, color='skyblue', edgecolor='black')
plt.title ('SPREAD OF SULFUR DIOXIDE CONTENT IN RED WINE')
plt.xlabel('FREE SULFUR DIOXIDE CONTENT IN RED WINE ')
plt.ylabel ('Frequency')
plt.show()
```

### The first scatter plot for a example, shows a possible correlation between alcoholic content and the fixed acidity in red wine, a cluster of dots can be seen in the middle, thus forming a pattern. Some outliers can be seen at the extreme left and right of the plot, showing a strong relationship between the two sets of data.
### The second scatter plot shows a relationship between residual sugar and sulphates in the red wine.
